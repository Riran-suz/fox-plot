```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['lines.markersize'] = 0.3
```

```python
fox = pd.read_csv("kitsune.csv")
_, ax = plt.subplots(figsize=(18,12))
ax.scatter(fox.x, fox.y)
ax.scatter(fox.x+150, fox.y)
ax.set_xlim(0,360)
ax.set_ylim(0,240)
```
